# FilterBasedTransformations: 
- smoothing: gaussian([4:16])
- noise: gaussian(0.01:0.1), speckle(0.01:0.1), salt, pepper, s&p

## Constraints:
- images are 2^8 pixel-depth

```python
from skimage import data
img = data.camera()
```

```python
gaussian_param = 16 
import skimage.filter as filters

smooth = filters.gaussian(img, gaussian_param)
plt.imshow(smooth,cmap='gray');
```

```python
noise_gaussian_param = 0.1
import skimage.util as util
out=util.random_noise(img, mode='gaussian', seed=None, clip=True, mean=0, var=noise_gaussian_param)
#plt.imshow(out,vmin=0,vmax=1, cmap='gray', interpolation='nearest');
plt.imshow(out, cmap='gray', interpolation='nearest');
```

```python
out=util.random_noise(img, mode='poisson', seed=None, clip=True)
plt.imshow(out, cmap='gray', interpolation='nearest');
```

```python
out=util.random_noise(img, mode='salt', seed=None, clip=True)
plt.imshow(out, cmap='gray', interpolation='nearest');
```

```python
out=util.random_noise(img, mode='pepper', seed=None, clip=True)
plt.imshow(out, cmap='gray', interpolation='nearest');
```

```python
out=util.random_noise(img, mode='s&p', seed=None, clip=True)
plt.imshow(out, cmap='gray', interpolation='nearest');
```

```python
noise_speckle_param = 0.1
import skimage.util as util
out=util.random_noise(img, mode='speckle', seed=None, clip=True, mean=0, var=noise_speckle_param)
#plt.imshow(out,vmin=0,vmax=1, cmap='gray', interpolation='nearest');
plt.imshow(out, cmap='gray', interpolation='nearest');
```
